# Análise de dados "Boston" em R
## Introdução
O Boston Housing Dataset é um conjunto de dados amplamente utilizado, principalmente por estar disponível gratuitamente, na análise estatística e aprendizado de máquina. O conjunto de dados contém informções sobre características de casas na cidade de Boston, mostra o número de quartos, proximidade de centros de empregos, número de quartos, etc.. O objetivo dos dados é prever o valor médio das casas ocupadas pelos proprietários.

## Objetivo
O objetivo deste trabalho é desenvolver e avaliar os modelos preditivos capazes de estimar o valor médio (medv) das casas no Boston Housing Dataset.

## Metodologia
Para conseguirmos estimar o medv utilizaremos diversas abordagens aprendidas na disciplina, os métodos utilizados serão: regressão linear, árvore de decisão e modelo de random forest. Finalmente, avaliaremos os modelos utilizando a validação cruzada e métricas de desempenho (RMSE, R²).

## Desenvolvimento
### Carregar os dados e analisar
Verificou-ses que não há dados nulos, verificamo o formato dos dados e assim poderemos continuar com a análise dos dados.

```r
library(MASS)
data("Boston")
head(Boston)
str(Boston)
sum(is.na(Boston))
```

### Instalação de pacotes
Optou-se pelo uso dos packages caret e randomForest, o caret para avaliação e comparação de modelos de aprendizado de máquina, enquanto o randomForest é justamente para termos a implementação otimizada do algoritimo que dá nome ao pacote.

```r
install.packages("caret")
install.packages("randomForest")
```

### Criação dos modelos
Utilizando as bibliotecas definimos então os três modelos, todos definidos segundo a variável train_control.

```r
library(caret)
#Definição do controle dos modelos
train_control = trainControl(method = "cv",number = 10, savePredictions = "final")

#Regressão linear
model_lm = train(medv ~ ., data = Boston, method = "lm", trControl = train_control)

# Árvore de decisão
model_tree <- train(medv ~ ., data = Boston, method = "rpart", trControl = train_control)

# Modelo de Random Forest
model_rf <- train(medv ~ ., data = Boston, method = "rf", trControl = train_control)
```

### Comparando os modelos
Finalmente utilizamos a validação Cruzada para comparar os modelos.

```r
## Visualizar resultados
model_lm$results
model_tree$results
model_rf$results

# Exemplo de erro quadrático médio (RMSE)
rmse_lm <- sqrt(mean((model_lm$pred$obs - model_lm$pred$pred)^2))
rmse_tree <- sqrt(mean((model_tree$pred$obs - model_tree$pred$pred)^2))
rmse_rf <- sqrt(mean((model_rf$pred$obs - model_rf$pred$pred)^2))

# Exemplo de correlação
cor_lm <- cor(model_lm$pred$obs, model_lm$pred$pred)
cor_tree <- cor(model_tree$pred$obs, model_tree$pred$pred)
cor_rf <- cor(model_rf$pred$obs, model_rf$pred$pred)

# Exibir resultados
print(paste("RMSE Linear Model: ", rmse_lm))
print(paste("RMSE Decision Tree: ", rmse_tree))
print(paste("RMSE Random Forest: ", rmse_rf))

print(paste("Correlation Linear Model: ", cor_lm))
print(paste("Correlation Decision Tree: ", cor_tree))
print(paste("Correlation Random Forest: ", cor_rf))
```

### Análise dos resultados
Os três modelos foram gerados e posteriormente avaliados através da validação cruzada e métricas de desempenho como o RMSE (erro quadrático médio), o R².
