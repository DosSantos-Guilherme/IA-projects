```python
## Pandas para manipular os dataframes
import pandas as pd

print('Setup completo!')
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dansbecker/melbourne-housing-snapshot")

print("Path to dataset files:", path)
```

```python
try:
  dataFrame = pd.read_csv(path + '/melb_data.csv')
except:
  print('failed to find file')
dataFrame
```

Criando uma variável nova, chamada TotalRooms dentro do dataframe, ela será a soma de Rooms com Bedroom2.

```python
dataFrame['TotalRooms'] = dataFrame['Rooms'] + dataFrame['Bedroom2']
dataFrame.head(10)
```

Analisando a correlação da nova variável

```python
correlation = dataFrame['TotalRooms'].corr(dataFrame['Price'])
print(correlation)
```

A correlação com a variavel price é de 0.49, é uma correlação média pra alta, quanto mais quartos existirem na casa, maior será o valor dela \(price\)
